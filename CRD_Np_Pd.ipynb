# Design of Experiment 
Completely Randomized Design (CRD)


To Create : ANOVA-Table of Completely Randomized Design (CRD) with the help of NUMPY and PANDAS only

```python
import numpy as np
import pandas as pd
```

```python
A=[55,49,42,21,52]
B=[61,112,30,89,63]
C=[42,97,81,95,92]
D=[169,137,169,85,154]
```

```python
Dic={'A':A,'B':B,'C':C,'D':D}
```

```python
Df=pd.DataFrame(Dic)
```

```python
Df
```

```python
dfm=Df.melt()
```

```python
dfm
```

```python
treatment=dfm.variable.nunique()
```

```python
treatment
```

```python
total=dfm.value.count()
```

```python
total
```

```python
df_treatment=treatment-1
```

```python
df_total=total-1
```

```python
df_error=df_total-df_treatment
```

```python
m_A=np.mean(A)
m_B=np.mean(B)
m_C=np.mean(C)
m_D=np.mean(D)
```

```python
mean_all=[m_A,m_B,m_C,m_D]
```

```python
m_total=np.mean(mean_all)
```

```python
m_total
```

```python
ss_total=0
for i in dfm.value:
  ss_total=ss_total+((i-m_total)**2)
```

```python
ss_total
```

```python
len=[len(A),len(B),len(C),len(D)]
```

```python
len
```

```python
mean_all
```

```python
m_total
```

```python
ss_treatment=0
for i in mean_all:
   for j in len:
      ss_treatment=ss_treatment+j*((m_total-i)**2)
      break
```

```python
ss_treatment
```

```python
ss_error=ss_total-ss_treatment
```

```python
mss_treatment=ss_treatment/df_treatment
mss_error=ss_error/df_error
```

```python
mss_error
```

```python
fcal=mss_treatment/mss_error
```

```python
fcal
```

```python
src_V=['Treatment','Error','Total']
df=[df_treatment,df_error,df_total]
ssc=[ss_treatment,ss_error,ss_total]
mss=[mss_treatment,mss_error,np.nan]
f=[fcal,np.nan,np.nan]
```

```python
anv={'Source of Varition':src_V,'Degree of Freedom':df,'Sum Square':ssc,'Mean Sum Square':mss,'F value':f}
```

```python
Anv=pd.DataFrame(anv)
```

```python
Anv
```

```python
ftab=3.06
```

**Conclusion :**

```python
if fcal<ftab:
  print('Null Hypothesis Accepted : all the treatments have same effect')
else:
  print('Null Hypothesis Rejected : atleast one treatment has different effect')

```
